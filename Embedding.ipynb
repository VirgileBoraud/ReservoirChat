```python
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
tokens = tokenizer.tokenize("This is an example of the bert tokenizer")
print(tokens)
```

```python
token_ids = tokenizer.convert_tokens_to_ids(tokens)
print(token_ids)
```

```python
token_ids =  tokenizer.encode("This is an example of the bert tokenizer")
print(token_ids)

tokens = tokenizer.convert_ids_to_tokens(token_ids)
print(tokens)
```

```python
import torch
from transformers import BertModel

model = BertModel.from_pretrained("bert-base-uncased")

# get the embedding vector for the word "example"
example_token_id = tokenizer.convert_tokens_to_ids(["example"])[0]
example_embedding = model.embeddings.word_embeddings(torch.tensor([example_token_id]))

print(example_embedding.shape)
```

```python
king_token_id = tokenizer.convert_tokens_to_ids(["king"])[0]
king_embedding = model.embeddings.word_embeddings(torch.tensor([king_token_id]))

queen_token_id = tokenizer.convert_tokens_to_ids(["queen"])[0]
queen_embedding = model.embeddings.word_embeddings(torch.tensor([queen_token_id]))

cos = torch.nn.CosineSimilarity(dim=1)
similarity = cos(king_embedding, queen_embedding)
print(similarity[0])
```

```python
similarity = cos(example_embedding, queen_embedding)
print(similarity[0])
```

## Trying on the Q&A

```python
from transformers import BertTokenizer

with open('Q&A_format.md', 'r', encoding='utf-8') as file:
    QA = file.read()

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
tokens = tokenizer.tokenize(QA)
print(tokens)
```

```python
tokens_ids = tokenizer.convert_tokens_to_ids(tokens)
# tokens_ids = tokenizer.encode(tokens) With the encode (adding begining and end of phrases, we exceed the limit of token indices sequence length which is of 512)
print(tokens_ids)
```

```python
from transformers import BertModel
import torch

model = BertModel.from_pretrained('bert-base-uncased')
inputs = tokenizer(QA, return_tensors='pt', max_length=512, truncation=True, padding=True)
outputs = model(**inputs)

# Extracting the last hidden state (the embeddings)
embeddings = outputs.last_hidden_state

# Displaying the shape of the embeddings
print(embeddings.shape)
print(embeddings)
```

```python
question = "What is an ESN?"
searchDocs = embeddings.similarity_search(question)
print(searchDocs[:4].page_content)
```
